These objects/asteroids can thus prove to be harmful. Hence, it is wise to know what is surrounding us and what can harm us amongst those. Thus, this dataset compiles the list of NASA certified asteroids that are classified as the nearest earth object.

Machine Learning

```python
from sklearn.model_selection import train_test_split
X = df.drop('hazardous', axis =1)
y = df.hazardous
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier  
from sklearn.ensemble import RandomForestClassifier
```

```python
model1=LogisticRegression()
model2=SVC(kernel='poly')
model3=KNeighborsClassifier(n_neighbors=3)
model4=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model5=RandomForestClassifier()
```

```python
model1.fit(X_train,y_train)
model2.fit(X_train,y_train)
model3.fit(X_train,y_train)
model4.fit(X_train,y_train)
model5.fit(X_train,y_train)
```

```python
pred1=model1.predict(X_test)
pred2=model2.predict(X_test)
pred3=model3.predict(X_test)
pred4=model4.predict(X_test)
pred5=model5.predict(X_test)
```

```python
# print("Logistic Regression Accuracy",model1.score(X_test,y_test))
# print("SVM Accuracy",model2.score(X_test,y_test))
# print("KNN Accuracy",model3.score(X_test,y_test))
# print("Decision Tree Accuracy",model4.score(X_test,y_test))
# print("Random Forest Accuracy",model5.score(X_test,y_test))
```

```python
a=model1.score(X_test,y_test)
b=model2.score(X_test,y_test)
c=model3.score(X_test,y_test)
d=model4.score(X_test,y_test)
e=model5.score(X_test,y_test)
```

```python
print("Logistic Regression Accuracy",a*100)
print("SVM Accuracy",b*100)
print("KNN Accuracy",c*100)
print("Decision Tree Accuracy",d*100)
print("Random Forest Accuracy",e*100)
```
