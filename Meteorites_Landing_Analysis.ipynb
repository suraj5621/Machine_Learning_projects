Meteorites are objects from space that have fallen to the Earth's surface. These objects provide valuable information about the composition and origins of our solar system, and the study of meteorites has contributed greatly to our understanding of the universe. This dataset contains information on meteorite landings that have been recorded throughout history. The data includes the date and location of each landing, the type of meteorite, and its size and mass. This dataset provides a comprehensive record of meteorite landings, allowing researchers to track the distribution of meteorites over time and space, and to better understand the nature and frequency of impacts on the Earth's surface. Whether you are a scientist, historian, or simply interested in space and astronomy, this dataset is a valuable resource for gaining a deeper understanding of meteorites and their role in shaping our planet.

```python
top_classes = df['recclass'].value_counts().head(20)

sns.barplot(x=top_classes.index, y=top_classes.values)
plt.title("Prevelance of Meteorite Classes")
plt.xlabel("Meteorite Class")
plt.ylabel("Number of Meteorites")
plt.xticks(rotation=90)
plt.show()
```

```python

g=sns.displot(df['year'])
g.fig.set_figwidth(15)
g.fig.set_figheight(10) 
plt.title("Distribution of Meteorites over Time")
plt.xlabel("Year")
plt.ylabel("Number of Meteorites")
plt.show()
```

```python
data_1950 = df[df['year'] > 1950]
data_1950_now = data_1950[data_1950['year']<2023]
```

```python
h=sns.displot(data_1950_now['year'])
h.fig.set_figwidth(15)
h.fig.set_figheight(10) 
plt.title("Distribution of Meteorites over Time")
plt.xlabel("Year")
plt.ylabel("Number of Meteorites")
plt.show()
```
